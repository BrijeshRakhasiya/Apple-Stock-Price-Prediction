```python
import yfinance as yf

ticker = 'AAPL'
start_date = '2024-07-10'
end_date = '2025-07-10'

# Download historical data
data = yf.download(ticker, start=start_date, end=end_date)

# Save to CSV
data.to_csv('AAPL_10Jul2024_to_10Jul2025.csv')
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.graph_objects as go
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("AAPL_10Jul2024_to_10Jul2025.csv")
df.head()
```

```python
# Clean the DataFrame -> remove first two rows and rename 'Price' column to 'Date'
df = df.iloc[2:].copy()
df = df.rename(columns={'Price': 'Date'})
df.reset_index(drop=True, inplace=True)
df.head()
```

- predict the stock prices of apple for the next 5 days

**AutoTS is an open-source Python library designed for automated time series forecasting. It automatically tests and selects the best models and parameters for forecasting tasks, making it easier to predict future values in time series data (like stock prices) without manual model selection and tuning.**

```python
from autots import AutoTS

model = AutoTS(forecast_length = 5 , frequency = 'infer' , ensemble = 'simple')
model = model.fit(df , date_col ='Date' , value_col='Close', id_col=None)
prediction = model.predict()
```

```python
forecast = prediction.forecast
print(forecast)
```
